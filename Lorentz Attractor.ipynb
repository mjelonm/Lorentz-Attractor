```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from mpl_toolkits.mplot3d import Axes3D

def Lorentz (sys, t):
    x = sys[0]
    y = sys[1]
    z = sys[2]
    
    sigma = 10.0
    r = 28.0
    b = 8.0/3.0
    
    dx = sigma * (y - x)
    dy = - x * z + r * x - y
    dz = x * y - b * z
    return dx, dy, dz

init = [1.0, 2.0, 3.0]
t = np.arange(0.0, 60.0, 0.01)

solution = odeint(Lorentz, init, t)

fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111, projection='3d')
ax.plot(solution[:,0], solution[:,1], solution[:,2])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Lorentz Attractor')
```
